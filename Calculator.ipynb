```python
class BasicCalc():
  def __init__(self,num1,num2):
    self.num1=num1
    self.num2=num2

  #addition
  def add(self):
    sum=self.num1+self.num2
    print(f"sum: {sum}")

  #subtraction
  def sub(self):
    diff=self.num1-self.num2
    print(f"Difference: {diff}")

  #multiplication
  def mul(self):
    prod=self.num1*self.num2
    print(f"product: {prod}")

  #division
  def div(self):
    division=self.num1/self.num2
    print(f"Division: {division}")

  #floor division
  def floordiv(self):
    floor_division=self.num1 // self.num2
    print(f"Floor Division: {floor_division}")
```

```python
#child class
import math

class AdvCalc(BasicCalc):
  def __init__(self, num1, num2):
    BasicCalc.__init__(self,num1,num2)
    self.num1 = num1
    self.num2 = num2

  #power
  def pow(self):
    power=self.num1 ** self.num2
    print(f"power: {power}")

  #modulus
  def mod(self):
    modulus=self.num1 % self.num2
    print(f"Modulus: {modulus}")

  #square root
  def sqrt(self):
    square_rt=self.num1 ** 0.5
    print(f"square root: {square_rt}")

  def log(self):
    log_cal=math.log(self.num1, self.num2) # Logarithm (base b): Calculates the logarithm of the first number with base b.
    print(f"Logarithm: {log_cal}")

  #greatest common divisor
  def gcd(self):
    gcd_cal=math.gcd(self.num1, self.num2)
    print(f"GCD: {gcd_cal}")



```

```python
while(True):
    print("Please enter 'a' to use basic calculator and 'b' for advanced calculator, or 'q' to quit ")
    choice=input("Enter your choice: ")
    if choice == 'q':
        break
    numb1=int(input("Enter the first number: "))
    numb2=int(input("Enter the second number: "))

    if (choice == 'a'):
        basic_choice = int(input("Enter 1 for Addition, 2 for Subtraction, 3 for Multiplication, 4 for Division, 5 for Floor Division: "))
        basic = BasicCalc(numb1,numb2)
        if basic_choice == 1:
            result = basic.add()
        elif basic_choice == 2:
            result = basic.sub()
        elif basic_choice == 3:
            result = basic.mul()
        elif basic_choice == 4:
            result = basic.div()
        elif basic_choice == 5:
            result = basic.floordiv()

    elif (choice == 'b'):
        adv_choice = int(input("Enter 1 for Power, 2 for modulus, 3 for square root, 4 for logarithm, 5 for GCD, 6 for Addition, 7 for Subtraction, 8 for Multiplication, 9 for Division, 10 for Floor Division: "))
        adv = AdvCalc(numb1,numb2)

        if adv_choice == 1:
            result = adv.pow()
        elif adv_choice == 2:
            result = adv.mod()
        elif adv_choice == 3:
            result = adv.sqrt()
        elif adv_choice == 4:
            result = adv.log()
        elif adv_choice == 5:
            result = adv.gcd()
        elif adv_choice == 6:
            result = adv.add()
        elif adv_choice == 7:
            result = adv.sub()
        elif adv_choice == 8:
            result = adv.mul()
        elif adv_choice == 9:
            result = adv.div()
        elif adv_choice == 10:
            result = adv.floordiv()





```
